```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data
train_X = train_df.drop(['loan_status', 'Unnamed: 0', 'index'], axis=1)
train_X_dummies = pd.get_dummies(train_X)

train_y = LabelEncoder().fit_transform(train_df['loan_status'])

train_X_dummies
```

```python
# Convert categorical data to numeric and separate target feature for testing data
test_X = test_df.drop(['loan_status', 'Unnamed: 0', 'index'], axis = 1)
test_X_dummies = pd.get_dummies(test_X)

test_y = LabelEncoder().fit_transform(test_df['loan_status'])

test_X_dummies
```

```python
# add missing dummy variables to testing set - missing debt_settlement_flag_Y column in testing data
print(f"Train: {train_X_dummies.shape}, Test: {test_X_dummies.shape}")

test_X_dummies["debt_settlement_flag_Y"]=0
        
print(f"Train: {train_X_dummies.shape}, Test: {test_X_dummies.shape}")
```

### Model Prediction:
I predict that the Random Forest Classifier model will perform better than the Logistic Regression overall because the Random Forest works with smaller subsets of data to build a strong classifier, whereas the Logistic Regression uses the dataset as a whole for its predictions

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(train_X_dummies, train_y)
print(f"Training Data Score: {classifier.score(train_X_dummies, train_y)}")
print(f"Testing Data Score: {classifier.score(test_X_dummies, test_y)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
RFclassifier = RandomForestClassifier(random_state=42).fit(train_X_dummies, train_y)
print(f'Training Score: {RFclassifier.score(train_X_dummies, train_y)}')
print(f'Testing Score: {RFclassifier.score(test_X_dummies, test_y)}')
```

```python
# Scale the data
scaler = StandardScaler().fit(train_X_dummies)
train_X_scaled = scaler.transform(train_X_dummies)
test_X_scaled = scaler.transform(test_X_dummies)
```

### Scaled Model Prediction:
I expect that the Linear Regression model will improve with the scaled data, but I would still expect the Random Forest model to perform better, as scaling may allow it to find more accurate trees with the subsets of data. 

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier = LogisticRegression()
classifier.fit(train_X_scaled, train_y)
print(f"Training Data Score: {classifier.score(train_X_scaled, train_y)}")
print(f"Testing Data Score: {classifier.score(test_X_scaled, test_y)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
RFclassifier = RandomForestClassifier(random_state=42).fit(train_X_scaled, train_y)
print(f'Training Score: {RFclassifier.score(train_X_scaled, train_y)}')
print(f'Testing Score: {RFclassifier.score(test_X_scaled, test_y)}')
```

### Model Results:
The Random Forest model did better with the un-scaled data than the Logistic Regression model, but the Logistic Regression model with the scaled data did the best overall. The Random Forest model did about the same with both the scaled and unscaled data, which I found surprising.
